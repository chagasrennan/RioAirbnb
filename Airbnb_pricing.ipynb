## Vamos separar entre baheiros compartilhados "shared" e privativos

Nan e 0 serão classificados como "shared" 

````python
def shared_bath(x):
    if 'shared' in x:
        return 1
    elif '0 baths' in x:
        return 1
    else:
        return 0
````

````python
listings.bathrooms_text.fillna('shared', inplace=True)
listings['private_bath'] = listings.bathrooms_text.apply(lambda x: shared_bath(x))

listings.drop(['bathrooms_text'],axis=1,inplace=True)
````

### Vou clusterizar os bairros por latitude e longitude. 

Assim a informação de localização será mais útil. Teremos poucos clusters com muitos pontos

````python
from sklearn.cluster import KMeans
````

````python
X=listings[['id','latitude','longitude']]
X.head(10)
````

````python
K_clusters = range(5,20)
kmeans = [KMeans(n_clusters=i) for i in K_clusters]

Y_axis = listings[['latitude']]
X_axis = listings[['longitude']]

score = [kmeans[i].fit(Y_axis).score(Y_axis) for i in range(len(kmeans))]


plt.plot(K_clusters, score)
plt.xlabel('Número de clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
````

````python
kmeans = KMeans(n_clusters = 12, init ='k-means++')

kmeans.fit(X[['latitude','longitude']])

X['cluster_label'] = kmeans.fit_predict(X[['latitude','longitude']])

centers = kmeans.cluster_centers_ 

labels = kmeans.predict(X[['latitude','longitude']]) 

X.head(10)
````

````python
cluster_id = X[['id','cluster_label']]
cluster_id.set_index('id',inplace=True)
````

````python
fig, ax = plt.subplots(figsize=(20, 8))


neighb.plot(ax=ax,  color='white', edgecolor='black')

X.plot.scatter(x = 'longitude', y = 'latitude', c=labels, s=50, cmap='viridis',ax=ax)

ax.axis('off')
````

# Modelo Preditivo

````python
df = listings.set_index('id').join(cluster_id)

for i in range(df.cluster_label.max() -1 ):
    df[f'cluster{i}'] = df.cluster_label.apply(lambda x: 1 if x==i else 0)
    

price_cut = df.price.quantile(.95)

df = df[(df.price<=price_cut)&(df.price>0)].copy()

df.price = np.log(df.price)

df.drop(['neighbourhood_cleansed','latitude','longitude','cluster_label'],axis=1, inplace=True)
````

````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn import metrics

scaler_features = StandardScaler()
scaler_coef = StandardScaler()

df[[c for c in df.columns if c!='price']] = scaler_features.fit_transform(df[[c for c in df.columns if c!='price']])

df[['price']] = scaler_coef.fit_transform(df[['price']])

Y = df['price']

X = df[[c for c in df.columns if c!='price']]

X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size=0.3, random_state=42)


regressor = LinearRegression()
regressor.fit(X_train,Y_train)

y_pred= regressor.predict(X_test)

y_pred_train= regressor.predict(X_train)
````

````python
explained_variance=metrics.explained_variance_score(Y_test, y_pred)
mean_absolute_error=metrics.mean_absolute_error(Y_test, y_pred) 
mse=metrics.mean_squared_error(Y_test, y_pred) 
median_absolute_error=metrics.median_absolute_error(Y_test, y_pred)
r2=metrics.r2_score(Y_test, y_pred)

print('explained_variance: ', round(explained_variance,4))    
print('r2: ', round(r2,4))
print('MAE: ', round(mean_absolute_error,4))
print('MSE: ', round(mse,4))
print('RMSE: ', round(np.sqrt(mse),4))

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,6), sharex=True, sharey=True)

ax1.scatter(y_pred_train, Y_train, alpha=0.4)
ax1.set_title('Dataset de treino')

ax2.scatter(y_pred, Y_test, alpha=0.4, c='g')
ax2.set_title('Dataset de Teste')
````
